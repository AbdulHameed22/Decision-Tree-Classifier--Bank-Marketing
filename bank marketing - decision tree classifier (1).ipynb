```python
import zipfile
import os

# Path to the downloaded zip file for bank-additional
zip_file_path_additional = "C:\\Users\\NHI652\\Downloads\\bank+marketing\\bank-additional.zip"

# Directory where you want to extract the files for bank-additional
extracted_folder_path_additional = "C:\\Users\\NHI652\\Downloads\\bank+marketing\\bank-additional"

# Extract the contents of the zip file for bank-additional
with zipfile.ZipFile(zip_file_path_additional, 'r') as zip_ref_additional:
    zip_ref_additional.extractall(extracted_folder_path_additional)

# Check the contents of the extracted folder for bank-additional
extracted_files_additional = os.listdir(extracted_folder_path_additional)
print("Extracted files for bank-additional:", extracted_files_additional)
```

```python
# Load the dataset from the extracted folder for bank
file_path_bank = os.path.join(extracted_folder_path, "bank.csv")
data_bank = pd.read_csv(file_path_bank, sep=';')

# Display the first few rows of the dataset for bank
print(data_bank.head())
```

```python
# Display information about the dataset
print(data_bank.info())

# Check for missing values
print(data_bank.isnull().sum())

# Encode categorical variables
data_bank_encoded = pd.get_dummies(data_bank)

# Split the data into features (X) and target variable (y)
X_bank = data_bank_encoded.drop('y_yes', axis=1)  # Features
y_bank = data_bank_encoded['y_yes']  # Target variable
```

```python
# Split the data into training and testing sets for bank
X_train_bank, X_test_bank, y_train_bank, y_test_bank = train_test_split(X_bank, y_bank, test_size=0.2, random_state=42)
```

```python
# Create a decision tree classifier
clf_bank = DecisionTreeClassifier(random_state=42)

# Train the classifier for bank
clf_bank.fit(X_train_bank, y_train_bank)
```

```python
# Make predictions on the test set for bank
y_pred_bank = clf_bank.predict(X_test_bank)
```

```python
# Evaluate the model for bank
accuracy_bank = accuracy_score(y_test_bank, y_pred_bank)
conf_matrix_bank = confusion_matrix(y_test_bank, y_pred_bank)
class_report_bank = classification_report(y_test_bank, y_pred_bank)

print("Accuracy for bank:", accuracy_bank)
print("Confusion Matrix for bank:\n", conf_matrix_bank)
print("Classification Report for bank:\n", class_report_bank)
```
